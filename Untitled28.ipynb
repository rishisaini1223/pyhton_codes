decorater are the high order function which takes another function as an argument which enhance the function of the existing without changing the code of original funnction.

```python
# decoraters
def add_fun(fun):
  def add_func():
    print('this is the msg')
    fun()
  return add_func

@add_fun
def hello():
  print('hello from hello function')

hello()
```

```python
def my_decorator(gift):
  def wrapper():
    print('adding some details')
    gift()
    print('addingsome feature')
  return wrapper

@my_decorator
def gift():
  print('this is book')

gift()
```

```python
# write a decorator function that prints 'function starting' before excuting a function.
def fun_2(fun):
  def print_f():
    print('function starting')
    fun()
  return print_f

@fun_2
def fun_1():
  print('this is fun_1')

fun_1()
```

```python
# create a decorator to check the num  is even before the calling function.
def dec_fun(fun):
  def even():
    n = int(input('enter a num'))
    if n%2==0:
      print('number is even')
    else:
      print('num is not even')
    fun()
  return even


@dec_fun
def fun_3():
  print('this is not decorator function')

fun_3()
```

```python
# decorator to check if number is even before calling the function

def check_num(fun):
    def func(num):
        if num % 2 == 0:
            print("Number is even")
            fun(num)   # call original function only if even
        else:
            print("Number is odd, function will not run")
    return func


@check_num
def display(n):
    print(f"This number is {n}")


display(5)
display(8)
```

```python
#
is_permission = True
def permission_access(fun):
  def wrapper():
    if is_permission:
      fun()
    else:
      print('permission Denied')
  return wrapper

@permission_access
def data():
  print('this is secert data')

data()
```

Iterator is an object that allows you to traverse through elements of a collection (like lists, tuples, dictionaries, etc.) one item at a time. It maintains the current state of the iteration and provides a memory-efficient way to process data, especially large datasets or infinite sequences, as it only loads one item into memory at a time

```python
li = [1,2,3,4,5,6,7]
# here we have made the normal list to the  iterables using the iter function.
iterable = iter(li)
# next keyword to access each element from the iterables.
next(iterable)
next(iterable)
next(iterable)
```
