When we have to operate any operation on the file then we must have to open that file fisrt.
syntax of opening a file ->
f = open('filename','mode')

'r' -> read mode

file = open('name.txt','r')

```python
file = open('name.txt','r')
content = file.read()
print(content)
```

```python
# for read file
try:
  file = open('nae.txt','r')
except FileNotfounderror:
  print('Please enter the valid file address')
else:
  content = file.read()
  print(content)
finally:
  file.close()
  print('File is closed')
```

```python
# for read single line of the file we use readline()
# Readlines -> it return the list of lines.
try:
  file = open('nae.txt','r')
except FileNotfounderror:
  print('Please enter the valid file address')
else:
  content = file.readline()
  print(content)
finally:
  file.close()
  print('File is closed')
```

```python
#in write mode it grand the access to write the file.
# if the file doesn't exist then it creates a new file otherwise it overwrite the file.
file = open('detail.txt','w')
```

```python
# for write the list of lines then we use writelines()
file.writelines(['this is  first line.\n','this is second line.\n''this is third line.\n'])
```

```python
# for adding new content in file , we use append mode - 'a'.

f = open('name.txt','a')
f.write('\nthis is a new line')
f.close()
```

```python
# we can perform  append and read on same time by using 'a+'
f = open('names.txt','a+')  # append, read operation -> file must exist
f.write('\n this is a line from the write mode')

# seek method is used to seek the pointer to any  where.
f.seek(8)
print(f.read())
f.close()
```

```python
# 'r+' -> read and write both. the file must exist.
f = open('names.txt','r+')  # append, read operation -> file must exist
f.write('\n this is a line from the write mode')

# seek method is used to seek the pointer to any  where.
f.seek(0)
print(f.read())
f.close()
```

```python
# with open('filename','mode') as file_alis

# with statement is used to open a file as well close the file at the same time. whenever we open a file in to the with statement. so the execution on file
# will occur inside the with statement block.
# after excution file is closed automatially to release the resourse
with open('name.txt','r') as n:
  print(n.read())
```
