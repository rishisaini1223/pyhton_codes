```python
# write a program to check weather a number is prime or not
def is_prime(n):
  for i in range(2,n//2+1):
    if n%i == 0:
      return False

  return True
n = int(input('enter num: '))
if is_prime(n):
  print('it is a prime ')
else :
  print('not prime')
```

```python
# variable length argument s
def add(*args):
  total = 0
  for el in args:
    total+=el
  return total

print(add(3,4,3,2,3))
```
