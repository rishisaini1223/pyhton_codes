```python
def fact(n):
  if n==0 or n== 1:
    return 1
  else:
    return n*fact(n-1)
print(fact(5))
```

```python
def fab(n):
  if n == 0:
    return 0
  elif n== 1:
    return 1
  else:
    return fab(n-1)+fab(n-2)

def fab_series(n):
  for i in range(n):
    print(fab(i),end=' ')
print(fab_series(8))
```

```python
# Write a program to find common elements between two lists using sets.
l1 = [1,2,3,4]
l2 = [3,4,5,6]
s1 = set(l1)
s2 = set(l2)
print(s1.intersection(s2))
```

```python
# Write a program to check whether one set is a subset of another.
s1 = {1,2,3,4}
s2 = {3,4}
if (s2.issubset(s1)):
  print('it is subset')
else:
  print('it is not a subset')
```

```python
# Write a program to find the maximum and minimum element in a set.
s = {1,4,6,3,8}
max=0
min = float('inf')
for el in s:
  if el > max:
    max = el
print('max value is =',max)
for el in s:
  if el < min:
    min = el
print('min value is =',min)
```

```python
# Write a program to count vowels in a string using a set.
def count_vovel(s):
  count = 0
  vovel = {'a','e','i','o','u'}
  for ch in s.lower():
    if ch in vovel:
      count += 1
  return count

string = 'akskdjfeia'
print(count_vovel(string))
```
