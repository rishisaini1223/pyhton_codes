generator-  These are defined like normal functions but use the yield keyword instead of return. When called, they return a generator object and the code within does not execute until iteration begins.

```python
def simple_generator():
    yield 1
    yield 2
    yield 3

# Using the generator function
gen = simple_generator()
for value in gen:
    print(value)
```

```python
# write a program to print odd number in range 1 to 1000
# using generator
def gen():
  for i in range(1,1000):
    if i%2 !=0:
      yield i
g = gen()
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```
