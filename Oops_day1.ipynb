```python
 # Here we would make a student class.
class student:
  # by default vishal
  # class attribute - this  class attribute is same for all the objects or instance
  name = 'vishal'
  # constructor is automatically called when we create an obj  by giving one property or argument.
  def __init__(self, location):
    self.location = location

# here we have created a student object
obj1 = student('jaipur')
print(obj1.location)

# here we have accessed the object's attribute
print(obj1.name)
```

```python
class teacher():
  def __init__(self,name,subject):
    self.name = name
    self.subject = subject

t1 = teacher('rahul','hindi')
print(t1.name)
print(t1.subject)
```

```python
class classteacher(teacher):
  def __init__(self,name,subject,section):
    super().__init__(name,subject)
    self.section = section
  def display_detail(self):
    print(f'Name: {self.name}')
    print(f'subject name : {self.subject}')
    print(f'section: {self.section}')

  c1 = classteacher('ran','hindi','a')
  c1.display_detail()
```

```python
# multilevel inheritance
class grandfather:
  def grandfather_prop(self):
    print('grandfather property')
class father(grandfather):
  def father_prop(self):
    print('father propeerty')
class child(father):
  def child_prop(self):
    print('child property')

child = child()
child.father_prop()
child.grandfather_prop()
child.child_prop()
```

```python
class mother:
  def house_management(self):
    print('house management')
class father:
  def finance_management(self):
    print('fin management')

class child(mother, father):
  pass
child = child()
child.finance_management()
child.house_management()
```
