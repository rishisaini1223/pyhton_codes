```python
# 1. Print numbers from 1 to n using recursion.
def numbers(n):
  if n == 10:
    return
  print(n,end =" ")
  numbers(n + 1)
numbers(1)
```

```python
# 2. Print numbers from n to 1 using recursion.
def numbers(n):
  if n == 0:
    return
  print(n,end =" ")
  numbers(n - 1)
numbers(10)
```

```python
 # 3. Find the sum of first n natural numbers using recursion.
def sum_of_numbers(n):
  if n == 0:
    return 0
  return n + sum_of_numbers(n - 1)
sum_of_numbers(10)
```

```python
# 4. Find the factorial of a number using recursion.
def factorial(n):
  if n == 0:
    return 1
  return n * factorial(n - 1)
factorial(5)
```

```python
# 5. Find the power of a number a^b using recursion.
def power(a,b):
  if b == 0:
    return 1
  return a * power(a,b - 1)
power(2,3)
```

```python
# 6. Find the sum of digits of a number using recursion.
def sum_of_digits(n):
  if n == 0:
    return 0
  return n % 10 + sum_of_digits(n // 10)
sum_of_digits(523)
```

```python
# 7. Reverse a number using recursion.
def reverse(num, rev =0):
  if num  == 0:
    return rev
  return reverse(num // 10, rev*10+num %10)
reverse(123)
```

```python
# 8. Check whether a number is a palindrome using recursion.
def pallindrome(num):
  def reverse(n, rev =0):
    if n == 0:
      return rev
    return reverse(n//10, rev * 10 + n % 10)
  if num < 0:
    return
  if num == reverse(num):
    print("this is a pallindrome")
  else:
    print("this is not a pallindrome")
pallindrome(121)
```

```python
# 9. Find the nth Fibonacci number using recursion.
def fibonacci(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)
fibonacci(10)
```

```python
# 10. Print Fibonacci series up to n terms using recursion.
def print_fibonacci(n, a=0, b=1):
  if n == 0:
    return
  print(a, end=" ")
  print_fibonacci(n - 1, b, a + b)
print_fibonacci(5)
```

```python
# 11. Count the number of digits in a number using recursion.
def count_digits(n):
  if n == 0:
    return 0
  return 1 + count_digits(n // 10)
count_digits(123)
```

```python
# 12. Find the product of digits of a number using recursion.
def pro_digits(n):
  if n == 0:
    return 1
  return n % 10 * pro_digits(n // 10)
pro_digits(123)
```

```python
# 13. Check whether a string is a palindrome using recursion.
def is_palindrome(s):
  if len(s) <= 1:
    return True
  if s[0] != s[-1]:
    return False
  return is_palindrome(s[1:-1])
word = 'naman'
if is_palindrome(word):
  print('this is a pallindrome')
else:
  print('this is not pallindrome')
```

```python
# 14. Find the length of a string using recursion.
def length_of_string(s):
  if s == '':
    return 0
  return 1 + length_of_string(s[1:])
length_of_string('naman')
```

```python
# 15. Reverse a string using recursion.
def reverse_string(s):
  if len(s) <= 1:
    return s
  return reverse_string(s[1:]) + s[0]
reverse_string('madam')
```

```python
#  16. Find the sum of all elements in a list using recursion.
def sum_of_list(l):
  if len(l) == 0:
    return 0
  return l[0] + sum_of_list(l[1:])
sum_of_list([1,2,3,4,5])
```

```python
# 17. Find the maximum element in a list using recursion.
def find_max(l):
  if len(l) == 1:
    return l[0]
  return max(l[0], find_max(l[1:]))
find_max([1,2,3,4,5])
```

```python
# 18. Find the minimum element in a list using recursion.
def find_min(l):
  if len(l) == 1:
    return l[0]
  return min(l[0], find_min(l[1:]))
find_min([1,2,3,4,5])
```

```python
# 19. Check whether a number is even or odd using recursion.
def is_even(n):
  if n == 0:
    return True
  if n == 1:
     return False
  return is_even(n - 2)
num = 17
if is_even(num):
  print("Even")
else:
  print("Odd")
```

```python
# 20. Calculate GCD of two numbers using recursion.
def gcd(a,b):
  if b == 0:
    return a
  return gcd(b,a%b)
gcd(12,6)
```
