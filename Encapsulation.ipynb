```python
# encapsulations
 # a protective shield that puts restrictions on accessing variables
 # and methods directly, and can prevent accidental or unauthorized modification of data
```

public, protected, and private are not strict keywords like in other languages (e.g., Java, C++) but are naming conventions used to control access to class members (variables and methods)

```python
class atm:
  def __init__(self):
    self.name= 'vishal'
    self.__pin = 2343 # protected attribute
    # getter = it is use to get the private values.
  def show_var(self):
    print(self.__pin)
    # setter = setter function is used to set tha values of the private and restricted values.
  def change_pin(self,new_pin):
    self.__pin = new_pin

card = atm()
print(card.show_var())
card.change_pin(4554)
print(card.show_var())
```
